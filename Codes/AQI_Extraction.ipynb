```python
def to_unix(dtstr):
    return int(time.mktime(datetime.strptime(dtstr, '%d/%m/%Y').timetuple()))

START_DATE = '01/01/2023'
END_DATE = '24/11/2025'
start_unix = to_unix(START_DATE)
end_unix = to_unix(END_DATE)

results = []

for city in CITIES:
    url = f"http://api.openweathermap.org/data/2.5/air_pollution/history?lat={city['lat']}&lon={city['lon']}&start={start_unix}&end={end_unix}&appid={API_KEY}"
    try:
        r = requests.get(url)
        r.raise_for_status()
        data = r.json()
        for entry in data.get('list', []):
            row = {
                'city': city['name'],
                'date': datetime.utcfromtimestamp(entry.get('dt')).strftime('%Y-%m-%d %H:%M:%S'),
                'aqi': entry['main']['aqi'],
                'co': entry['components'].get('co', None),
                'no': entry['components'].get('no', None),
                'no2': entry['components'].get('no2', None),
                'o3': entry['components'].get('o3', None),
                'so2': entry['components'].get('so2', None),
                'pm2_5': entry['components'].get('pm2_5', None),
                'pm10': entry['components'].get('pm10', None),
                'nh3': entry['components'].get('nh3', None),
                'lat': city['lat'],
                'lon': city['lon']
            }
            results.append(row)
        print(f"Success: {city['name']} - {len(data.get('list', []))} records")
    except Exception as e:
        print(f"Error: {city['name']} - {str(e)}")
        time.sleep(1)


df = pd.DataFrame(results)
df.to_csv('indian_cities_air_quality_full.csv', index=False)
print(f"Saved {len(df)} air quality records for 50 top Indian cities.")
```

```python
# Step 3: Extract hour, date, month, and assign Indian season & quarter
df['hour'] = df['date'].dt.hour
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.date  # for daily grouping

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Summer'
    elif month in [6, 7, 8, 9]:
        return 'Monsoon'
    elif month in [10, 11]:
        return 'Post-monsoon'
    return 'Unknown'

df['season'] = df['month'].apply(get_season)

def assign_quarter(hour):
    if 0 <= hour < 6:
        return 'Q1'
    elif 6 <= hour < 12:
        return 'Q2'
    elif 12 <= hour < 18:
        return 'Q3'
    else:
        return 'Q4'

df['interval'] = df['hour'].apply(assign_quarter)

# Step 4: Aggregate by city, date, month, season, interval (mean of each value)
agg_dict = {
    'aqi': 'mean',
    'co': 'mean',
    'no': 'mean',
    'no2': 'mean',
    'o3': 'mean',
    'so2': 'mean',
    'pm2_5': 'mean',
    'pm10': 'mean',
    'nh3': 'mean',
    'lat': 'first',
    'lon': 'first'
}

df_agg = df.groupby(['city', 'day', 'month', 'season', 'interval'], as_index=False).agg(agg_dict)
df_agg.rename(columns={'day': 'date'}, inplace=True)

# Step 5: Save to a new compact CSV file
df_agg.to_csv('indian_cities_air_quality_aggregated.csv', index=False)
print(f"Aggregated file saved as 'indian_cities_air_quality_aggregated.csv'")
```
