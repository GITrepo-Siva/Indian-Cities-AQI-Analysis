```python
# Visual Python: Data Analysis > Import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df1 = pd.read_excel('./AQI_50India.xlsx')
df1.head()
```

```python
df2 = pd.read_excel('./indiancities_weather_q1_q4_openmeteo.xlsx')
df2.head()
```

```python
#
```

```python
df1['date'] = pd.to_datetime(df1['date'])
df2['date'] = pd.to_datetime(df2['date'])
```

```python
# Merging on city, date, interval
merged_df = pd.merge(
    df1,
    df2,
    on=['city', 'date', 'interval'],
    how='inner'   
)
```

```python
# Save final merged file
merged_df.to_excel("merged_output.xlsx", index=False)

merged_df.head()
```

```python
## Finding the null values if any
```

```python
# Visual Python: Data Analysis > Data Info
pd.DataFrame({'Null Count': merged_df.isnull().sum(), 'Non-Null Count': merged_df.notnull().sum()})
```
